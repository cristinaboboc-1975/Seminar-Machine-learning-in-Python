# Curățarea bazei de date

```python
def drop_columns_with_high_nulls(df, threshold_percentage):

    # Calculam numarul maxim de valori lipsa admise pe baza unei limite definite
    threshold = len(df) * (threshold_percentage / 100)

    # Obtinem lista variabilelor ce trebuie inlaturate din analiza
    dropped_columns = df.columns[df.isnull().sum() > threshold]

    # stergem variabilele cu multe valori lipsa
    df_cleaned = df.drop(dropped_columns,  axis=1)

    return df_cleaned, dropped_columns


threshold_percentage = 10  # Poate fi schimbată această limită - noi am considerat ca un număr de peste 30 valori lipsă este prea ridicat (adică un procent de peste 3,5%)

# Curățarea tabelului prin eliminarea coloanelor cu multe valori lipsă și obținerea listei variabilelor eliminate
df_cleaned, dropped_columns = drop_columns_with_high_nulls(df, threshold_percentage)

# Afisarea tabelului curățat
print("Cleaned DataFrame:")
print(df_cleaned)

# Display the list of columns that were dropped
print("\nDropped Columns:")
print(dropped_columns)

df = df_cleaned
#df.info()
```

Nu a fost inlaturata nici o variabila

```python
# Care variabile sunt numerice si care sunt categoriale?

# Selectarea caracteristicilor categoriale
cate_data = df.select_dtypes("object").columns
print("Categorical variables = ",cate_data.shape)

# Selectarea caracteristicilor numerice
num_data = df.select_dtypes(include=["int64", "float64"]).columns
print("Numerical variables = ",num_data.shape)
```

```python
#Inlocuirea valorilor lipsa cu media
for col in num_data:
    df[col].fillna(df[col].mean(), inplace=True)
```
